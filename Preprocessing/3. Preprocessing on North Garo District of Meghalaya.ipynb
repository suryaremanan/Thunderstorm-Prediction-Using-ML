We took a smaller dataset to reduce the bias. We took North Garo as our location which is very prone to lightning

```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('north garo.csv')
```

```python
df['date']=pd.to_datetime(dict(year=df.YEAR, month=df.MO, day=df.DY))
```

```python
df.head()
```

```python
df['date1']=pd.to_datetime(df.date)
```

```python
df['hr'] = pd.to_datetime(df.HR, format='%H')
```

```python
df['Time'] = pd.to_datetime(df['hr']).dt.time
```

```python
df["Time"]= pd.to_datetime(df["hr"])
```

```python
df['timestamp']=pd.to_datetime(df.date) + df.HR.astype('timedelta64[h]')
```

```python
df.head()
```

```python
df.drop(['MO', 'DY','HR','hr','Time'], axis=1, inplace=True)
```

```python
df.drop(['date', 'date1'], axis=1, inplace=True)
```

```python
df.head()
```
